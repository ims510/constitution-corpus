```python
"""
This script combines the two CSV files containing the Moldovan and Romanian news articles, scraped using the scripts
moldovan_scraping.py and romanian_scraping.py, respectively. 
The two CSV files are combined into a single dataset.
"""


import pandas as pd

csv_files = ['/Users/madalina/Documents/M1TAL/outils_trait_corpus/constitution-corpus/data/clean/moldovan_news.csv', '/Users/madalina/Documents/M1TAL/outils_trait_corpus/constitution-corpus/data/clean/romanian_news.csv']
df_csv_append = pd.DataFrame()
 
# append the CSV files
for file in csv_files:
    df = pd.read_csv(file)
    df_csv_append = df_csv_append.append(df, ignore_index=True)
 
dataset = df_csv_append
dataset
```
